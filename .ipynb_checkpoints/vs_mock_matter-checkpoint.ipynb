```python
# define cov Legendre transform function

def LegendreTrans_Cov(l,p,lp=None,mu_max=1.,Nk=500,Nmu=50):
  '''
  Returns the l'th multipole of |P(k,mu), where P(k,mu) is a 
  vector of length Nk*Nmu. Returns a vector of length Nk.
  '''
  if lp is None: lp = l
  # n = self.Nk ; m = self.Nmu
  n = Nk; m = Nmu
  # mu = self.mu.reshape((n,m))[0]
  mu = np.linspace(0.,1.,50)
  # MU = np.tile(mu,Nk)
  # mu = self.mu.reshape((n,m))[0]
  p_reshaped = p.reshape((n,m))
  result = np.zeros(n)
  for i in range(n):
     integrand = (2*l+1)*(2*lp+1)*p_reshaped[i,:]*scipy.special.legendre(l)(mu)*scipy.special.legendre(lp)(mu)
     result[i] = scipy.integrate.simps(integrand,x=mu)
  return result
def get_pars(path,EPT=True):
    file1 = open(path, 'r')
    Lines = file1.readlines()

    # Strips the newline character
    param_labels = Lines[0].split()
    param_labels = param_labels[1:]
    params = Lines[1].split()
    
    p_names = ['b1','b2','bs','b3','alpha0','alpha2','alpha4','alpha6','SN0','SN2','SN4']
    LPT_params = []
    for pp in p_names:
        if pp in param_labels: LPT_params.append(float(params[param_labels.index(pp)]))
        elif pp=='b1': LPT_params.append(1.)
        else: LPT_params.append(0)
    LPT_params[0] -= 1
    if not EPT: return LPT_params
    
    EPT_params = np.zeros(len(p_names))
    EPT_params[-7:] = LPT_params[-7:]
    EPT_params[0] = LPT_params[0] + 1
    EPT_params[1] = LPT_params[1] + 8/21 * LPT_params[0]
    EPT_params[2] = LPT_params[2] - 2/7 * LPT_params[0]
    EPT_params[3] = LPT_params[3] # just leave b3.... not sure what to do with it
    
    return EPT_params
```

```python
def get_dp_dict(path):
    file1 = open(path, 'r')
    Lines = file1.readlines()

    # Strips the newline character
    param_labels = Lines[0].split()
    param_labels = param_labels[1:]
    params = Lines[1].split()
    try:
        p_names = ['dp0','dp2']
        p_dict = {}
        # get_params = []
        for pp in p_names:
            p_dict[pp] = (float(params[param_labels.index(pp)]))
    except:
        p_names = ['db','df']
        p_dict = {}
        # get_params = []
        for pp in p_names:
            if pp in param_labels: p_dict[pp] = (float(params[param_labels.index(pp)]))
            else: p_dict[pp]=0
        pars = get_pars('fits/rsd_matterfit_pkmu.minimum.txt')
        b1, b2, bs, b3, alpha0, alpha2, alpha4, alpha6, sn, sn2, sn4 = pars
        db, df = p_dict['db'], p_dict['df']
        p_dict['dp0'] = 2*b1*db+db**2 + 2/3 * (b1*df+db*df+f*db) + 7/35 * (df**2+2*f*df)
        p_dict['dp2'] = 4/3*(b1*df+db*df+f*db) +  20/35 * (df**2+2*f*df)
    for pp in ['SN0','SN2','SN4','Bshot']:
        try:
            p_dict[pp] = (float(params[param_labels.index(pp)]))
        except:
            continue

    return p_dict
```

```python
def get_dbdf(path,b1,f):
    file1 = open(path, 'r')
    Lines = file1.readlines()

    # Strips the newline character
    param_labels = Lines[0].split()
    param_labels = param_labels[1:]
    params = Lines[1].split()
    try:
        p_names = ['dp0','dp2']
        get_params = []
        for pp in p_names:
            get_params.append(float(params[param_labels.index(pp)]))
        dp0, dp2 = get_params
        # print(get_params,b1, f)
        def objective(vars):
            db, df = vars
            eq1 = 2*b1*db+db**2 + 2/3 * (b1*df+db*df+f*db) + 7/35 * (df**2+2*f*df) - dp0
            eq2 = 4/3*(b1*df+db*df+f*db) +  20/35 * (df**2+2*f*df) - dp2 
            return eq1**2 + eq2**2
        # Bounds for the variables
        bounds = ([-.5, -.5], [.5, .5])  # Lower and upper bounds for x and y
        # Initial guess
        initial_guess = [0, 0]

        # Use least_squares to solve the system with bounds
        # result = minimize(objective, initial_guess, bounds=bounds)
        result = minimize(objective, initial_guess)#, bounds=bounds)
        solution = result.x
        return solution
    except:
        db,df = get_dp_dict(path)['db'],get_dp_dict(path)['df'] 
        return db,df
```

```python
pars = get_pars('fits/rsd_matterfit_pkmu.minimum.txt')
gen_name = '%s/z%.3f/%s_matter'%('rsd',1.1,'m04_R30')
Cn = [0,-1,0,0]; Rcut = 30
temp_mark = mark(klin,plin,nk=200,Cn=Cn,R=Rcut,name=gen_name,N=5000,basedir='.')

def sigma2_R(temp_mark,R=-1):
    kint = temp_mark.kint
    plin_p = temp_mark.plin_p
    W_R_int = temp_mark.W_R(kint,R=R)
    sigma2_R = simps(kint**2*W_R_int*plin_p,x=kint)/(2*np.pi**2)
    return sigma2_R

def sigma2_RR(temp_mark,R=-1):
    kint = temp_mark.kint
    plin_p = temp_mark.plin_p
    W_R_int = temp_mark.W_R(kint,R=R)
    sigma2_RR = simps(kint**2*W_R_int**2*plin_p,x=kint)/(2*np.pi**2)
    return sigma2_RR

def sigmabar2_R(temp_mark,R=-1):
    kint = temp_mark.kint
    plin_p = temp_mark.plin_p
    W_R_int = temp_mark.W_R(kint,R=R)
    sigma2_R = simps(kint**2*W_R_int*plin_p * kint**2*R**2,x=kint)/(2*np.pi**2)
    return sigma2_R

def sigmabar2_RR(temp_mark,R=-1):
    kint = temp_mark.kint
    plin_p = temp_mark.plin_p
    W_R_int = temp_mark.W_R(kint,R=R)
    sigma2_RR = simps(kint**2*W_R_int**2*plin_p * kint**2*R**2,x=kint)/(2*np.pi**2)
    return sigma2_RR

def int_W_R(temp_mark,R=-1):
    kint = temp_mark.kint
    W_R_int = temp_mark.W_R(kint,R=R)
    return simps(kint**2*W_R_int,x=kint)/(2*np.pi**2)


R_plot = np.linspace(5,60,50)
s2_R_plot = [sigma2_R(temp_mark,R=RR) for RR in R_plot]
s2_RR_plot = [sigma2_RR(temp_mark,R=RR) for RR in R_plot]
sb2_R_plot = [sigmabar2_R(temp_mark,R=RR) for RR in R_plot]
sb2_RR_plot = [sigmabar2_RR(temp_mark,R=RR) for RR in R_plot]

s2_R_plot = np.array(s2_R_plot)
s2_RR_plot = np.array(s2_RR_plot)
sb2_R_plot = np.array(sb2_R_plot)
sb2_RR_plot = np.array(sb2_RR_plot)
```
