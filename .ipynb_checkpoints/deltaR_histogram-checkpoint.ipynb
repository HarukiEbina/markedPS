```python
import numpy as np
import json
from matplotlib import pyplot as plt
```

```python
# simid_list = ['AbacusSummit_base_c000_ph000','AbacusSummit_base_c000_ph001','AbacusSummit_base_c000_ph002']
simid_list = ['AbacusSummit_base_c000_ph002']

total_dict = {}

for simid in simid_list:
    filename = 'mocks/results/%s/z1.100/rsd_hist-Copy1.json'%simid
    with open(filename, "r") as json_file:
        data = json.load(json_file)
    for i,key in enumerate(data): 
        if key in total_dict: 
            if key[-4:]=='hist': total_dict[key]+=np.array(data[key])
        else: total_dict[key] = np.array(data[key])   
```

```python
fig,ax = plt.subplots(1,3,figsize=(10,2.6))

for i in range(3):
    # ax[i,0].stairs(total_dict['matter_%i_hist'%(10*(i+1))],total_dict['matter_%i_edges'%(10*(i+1))],density=True)
    # ax[i].hist(total_dict['matter_%i_edges'%(10*(i+1))][:-1], total_dict['matter_%i_edges'%(10*(i+1))], 
    #              weights=total_dict['matter_%i_hist'%(10*(i+1))],density=True,histtype='step',color='C0')
    ax[i].hist(total_dict['matter_%i_edges'%(10*(i+1))][:-1], total_dict['matter_%i_edges'%(10*(i+1))], 
                 weights=np.array(total_dict['matter_%i_hist'%(10*(i+1))])/np.max(total_dict['matter_%i_hist'%(10*(i+1))]),histtype='step',color='C0')
    # ax[i].set_xlim(-2,2)
    # ax[i].set_ylim(0,4)
    

    # ax[i,1].stairs(total_dict['LRG_%i_hist'%(10*(i+1))],total_dict['LRG_%i_edges'%(10*(i+1))])
    # ax[i].hist(total_dict['LRG_%i_edges'%(10*(i+1))][:-1], total_dict['LRG_%i_edges'%(10*(i+1))], 
    #              weights=total_dict['LRG_%i_hist'%(10*(i+1))],density=True,histtype='step',color='C1')
    ax[i].hist(total_dict['LRG_%i_edges'%(10*(i+1))][:-1], total_dict['LRG_%i_edges'%(10*(i+1))], 
                 weights=np.array(total_dict['LRG_%i_hist'%(10*(i+1))])/np.max(total_dict['LRG_%i_hist'%(10*(i+1))]),histtype='step',color='C1')
    ax[i].set_xlim(-1.3,1.3)
    ax[i].set_ylim(0,1.1)
    if i>0: ax[i].set_yticklabels([])
    else: ax[i].set_ylabel('$p(x)$')
    ax[i].set_xlabel('$\delta_{g,R}$')
    # ax[i].grid()
    
    ax[i].set_title('$R=%d \,h^{-1}\,{\\rm Mpc}$'%(10*(i+1)),x=.75,y=.85)

ax[0].plot([],[],'C0',label='$\delta_R$')
ax[0].plot([],[],'C1',label='$\delta_{g,R}$')
ax[0].legend(loc = 'upper left')

plt.tight_layout()
# plt.savefig('plots/deltaR_hist.png')
plt.show()
```
